##POS Tagger

```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from prettytable import PrettyTable
```

```python
content = "I am a super fan of assassin's creed"
```

```python
tagged = nltk.pos_tag(word_tokenize(content))
print(tagged)
```

```python
#Get only verbs
print([element for element, pos in tagged if pos in ['VS','VBD','VBG','VBN','VBP','VBZ']])
```

##HIDDEN MARKOV MODEL (Base Knowledge)

```python
content1 = "I love my work"
content2 = "you work well"
content3 = "I work hard today"
content4 = "the work is fun" 
```

```python
tagged1 = nltk.pos_tag(word_tokenize(content1))
print(tagged1)

tagged2 = nltk.pos_tag(word_tokenize(content2))
print(tagged2)

tagged3 = nltk.pos_tag(word_tokenize(content3))
print(tagged3)

tagged4 = nltk.pos_tag(word_tokenize(content4))
print(tagged4)
```

```python
tagged = nltk.pos_tag(word_tokenize("My work is fun"))
print(tagged)
```

```python
tagged = nltk.pos_tag(word_tokenize("I work hard"))
print(tagged)
```

```python
print([element for element, pos in tagged1 if pos in ['VS','VBD','VBG','VBN','VBP','VBZ']])

print([element for element, pos in tagged2 if pos in ['VS','VBD','VBG','VBN','VBP','VBZ']])

print([element for element, pos in tagged3 if pos in ['VS','VBD','VBG','VBN','VBP','VBZ']])

print([element for element, pos in tagged4 if pos in ['VS','VBD','VBG','VBN','VBP','VBZ']])
```
