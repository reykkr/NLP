## Tokenization

```python
sentence_1 ="I read a book. The book is on the table. I do not have a computer."
sentence_2 ="I have a computer on the table, and another computer in my bedroom. My computer is a real book."
sentence_3 ="What is a BOOK ? A BOOK ? I have a computer and I prefer it to any book."
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
tokens_1 = word_tokenize(sentence_1)
tokens_2 = word_tokenize(sentence_2)
tokens_3 = word_tokenize(sentence_3)
```

```python
print(tokens_1)
```

## Vectors creation

```python
nbBook = 0
nbComputer = 0
for token in tokens_1:
  if(token.lower() == "book"):
    nbBook+=1
  if(token.lower() == "computer"):
    nbComputer+=1

u = [nbBook,nbComputer]
```

```python
nbBook = 0
nbComputer = 0
for token in tokens_2:
  if(token.lower() == "book"):
    nbBook+=1
  if(token.lower() == "computer"):
    nbComputer+=1

v = [nbBook,nbComputer]
```

```python
nbBook = 0
nbComputer = 0
for token in tokens_3:
  if(token.lower() == "book"):
    nbBook+=1
  if(token.lower() == "computer"):
    nbComputer+=1

w = [nbBook,nbComputer]
```

```python
print(u)
print(v)
print(w)
```

## Cosine similarity

$$cos\theta = \frac{\vec{u} . \vec{v}}{\vert\vert \vec{u} \vert\vert \vert\vert\vec{v}\vert\vert}$$

```python
import math
def magnitude(vector):
  print("Vector dimension : ",len(vector))
  sum = 0
  for i in range(0,len(vector)):
    sum += vector[i]**2
  
  return math.sqrt(sum)
```

```python
def dotProduct(vector1, vector2):
  if(len(vector1)== len(vector2)):
    sum = 0
    for i in range(0,len(vector1)):
      sum += vector1[i] * vector2[i]
    return sum
  else:
   print("Error !")
```

```python
def cosineSimilarity(vector1, vector2):

  magnitude_vector_1 = magnitude(vector1)
  print("Magnitude vector 1 ",magnitude_vector_1)
  magnitude_vector_2 = magnitude(vector2)
  print("Magnitude vector 2 ",magnitude_vector_2)
  dot_product = dotProduct(vector1,vector2)
  print("Dot product vector 1 and vector 2 ",dot_product)
  return (dot_product / (magnitude_vector_1*magnitude_vector_2))
```

```python
#Text 1 / Text 2
cosineSimilarity(u,v)
```

```python
#Text 1 / Text 3
cosineSimilarity(u,w)
```
